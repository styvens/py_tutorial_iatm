```python
# visualization of neuroimage in Nilearn module. Este modilo tiene ya imageens procesadas
from nilearn import datasets

# default MNI152 template

anatomical_image=datasets.load_mni152_template()
# them, import ploting module

from nilearn import plotting
# to plot anatomical image
plotting.plot_anat(anatomical_image,title='mni template')




```

```python
#Teth t-contraste maps from Nilearn datasets module
localizer_dataset=datasets.fetch_localizer_button_task(
n_subjects=1,get_anats=True)

# anatomcal image filname

localizer_anat_filname=localizer_dataset.anats[0]
print(localizer_anat_filname)

# T-contrast map filname
localizer_tmap_filname=localizer_dataset.tmaps[0]
print(localizer_tmap_filname)






```

```python
plotting.plot_stat_map(localizer_tmap_filname,bg_img=localizer_anat_filname,threshold=3,
                       title="statistical map with threshols=3")
```

```python
plotting.plot_glass_brain(localizer_tmap_filname,threshold=3,plot_abs=False,colorbar=True,
                         title="statistical map with plot abs=False and threshols=3")
```

```python
# showing percentile thereshold
plotting.plot_stat_map(localizer_tmap_filname,display_mode='z',threshold=3,
                      cut_coords=5,colorbar=True)
```
